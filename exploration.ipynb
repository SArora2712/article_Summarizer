```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from app.summarizer import summarize

text = """
Artificial intelligence is transforming many industries.
Machine learning is a subset of artificial intelligence.
NLP allows machines to understand human language.
Text summarization is an important NLP task.
"""

summary = summarize(text, top_n=3)

wc = WordCloud(width=800, height=400, background_color='white').generate(" ".join(summary))

plt.figure(figsize=(10,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
print("ORIGINAL TEXT:\n", text)
print("\nSUMMARY:\n", " ".join(summary))
```
